## Swap the elemnts in the list that is the last number and first number

```python
## using temporvary variable
my_list=[1,2,3,4,5,6]
size=len(my_list)

temp=my_list[0]
my_list[0]=my_list[size-1]
my_list[size-1]=temp
print(my_list)
```

```python
my_list=[1,2,3,4,5,6]
my_list[0],my_list[-1]=my_list[-1],my_list[0]
print(my_list)
```

```python
my_list=[1,2,3,4,5,6]
start,*middle,end=my_list
print(start)
print(middle)
print(end)
my_list=[end,*middle,start]
print(my_list)
```

```python
my_list=[1,2,3,4,5,6]
first=my_list.pop(0)
last=my_list.pop(-1)

my_list.insert(0,last)
my_list.append(first)

print(
    my_list)
```

## swap Two elemnt in any postion of list

```python
my_list=[1,2,3,4,5,6]
pos1,pos2=1,3
first=my_list.pop(pos1)
last=my_list.pop(pos2-1)

my_list.insert(pos1,last)
my_list.insert(pos2,first)

print(my_list)
```

```python
my_list=[1,2,3,4,5,6]
print("Before swaping list is :",my_list)
pos1,pos2=1,3
my_list[pos1],my_list[pos2]=my_list[pos2],my_list[pos1]
print("After swaping the list is:",my_list)
```
