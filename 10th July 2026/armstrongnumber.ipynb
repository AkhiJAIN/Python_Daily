## FInding The Armstrong Number

```python
n=int(input("Enter the number:"))
temp=n
num=len(str(n))
total=0
while temp > 0:
    ld=temp % 10
    arm=ld**num
    total=total+arm
    temp=temp//10
if total==n:
    print("Yes ARM")
else:
    print("No Arm")
```

```python
n=153
num=str(n)
power=len(num)
total=0
for digit in num:
    total=total+int(digit) ** power
if total==n:
    print("Yes ARM")
else:
    print("No Arm")  
```
