```python
n = 12321

num = n
result = 0

while num > 0:
    ld = num % 10
    result = (result * 10) + ld
    num = num // 10


if n == result:
    print("Palindrome")
else:
    print("Not Palindrome")
```

## Plaindrome for the String

```python
s="madam"
reverse= s[::-1]
print(reverse)
if s== reverse:
    print("Palindrome")
else:
    print("No planidrome")
```

```python
s="akhilesh"
left=0
right=len(s)-1
flag=True
while left<right:
    if s[left]!=s[right]:
        flag=False
        break
    left+=1
    right-=1

if flag:
    print("Plaindrome")
else:
    ("No plaindrome")
```

```python
s = "madam"

left = 0
right = len(s) - 1

flag = True

while left < right:

    if s[left] != s[right]:
        flag = False
        break

    left += 1
    right -= 1


if flag:
    print("Palindrome")
else:
    print("Not Palindrome")
```
