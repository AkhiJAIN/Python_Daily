## check the prime number

the prime number is that which diveds by 1 or itself only then its prime number

```python
num = 5
count = 0

if num > 1:
    for i in range(1, num + 1):
        if num % i == 0:
            count += 1

    if count == 2:
        print("The number is prime")
    else:
        print("The number is not prime")
else:
    print("The number is not prime")
```
