## swap 2 numbers


```python
num1=20
num2=30

print("Number 1 before swaping is : ",num1)
print("Number 2 before swaping is :",num2)

temp=num1
num1=num2
num2=temp

print("value of num1 after swaping is:" ,num1)
print("value of num2 after swaping is :" ,num2)
```

```python
num1=input("Enter the number 1:")
num2=input("Enter the number 2:")
print("Number 1 before swaping is : ",num1)
print("Number 2 before swaping is :",num2)

temp=num1
num1=num2
num2=temp

print("value of num1 after swaping is:" ,num1)
print("value of num2 after swaping is :" ,num2)
```
