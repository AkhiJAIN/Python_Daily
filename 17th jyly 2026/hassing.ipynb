## HAsing problems


```python
n=[1,2,3,4,9,6,8]
m=[10,30,22,3,5,6,3,5,7,8]
for num in m:
    count=0
    for x in n:
        if x==num:
            count +=1
    print(count)
```

```python
n=[1,2,3,4,9,6,8]
m=[10,30,22,3,5,6,3,5,7,8]
hash_list=[0] * 11
for num in n:
    hash_list[num]+=1
for x in m:
    if x < 1 or x > 10:
        print(0)
    else:
        print(hash_list[x])

## Time complexity O[m+n]
## Space complexity O[11]~O[1]
```

```python
n=[1,2,3,4,9,6,8]
m=[10,30,22,3,5,6,3,5,7,8]
dict={}
for x in range(0,len(n)):
    if n[x] in dict:
        dict[n[x]]+=1
    else:
        dict[n[x]]=1
print(dict)
for num in m:
    if num < 0 or num > 10:
        print(0)
    elif num in dict:
        print(dict[num])
    else:
        print(0)
```
