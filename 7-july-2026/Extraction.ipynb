Question: The Number 
12345 is given to you and you should return the number of digits or count 
of digits in the given

```python
num=12345
n=num
count=0
while n > 0:
    count=count+1
    n=n//10
print( count)
```

```python
## susing the fuctions
def count_numbers(num):
    n=num
    count=0
    while n > 0:
        count=count+1
        n=n //10
    return count

result=count_numbers(1234)
print("The total numbers count is ," ,result)
```

```python
##using the logirithm 
from math import *
def countdigits(num):
    return int(log10(num)+1)
result=countdigits(1234)
print("The total numbers count is ," ,result)
```
