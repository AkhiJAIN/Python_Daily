## Sum of the element in the array

using the inbuilt fuction sum

```python
from builtins import sum
arr=[1,23,4,5,6,78,9]
print(sum(arr))
```

```python
arr=[1,23,4,5,6,78,9]
total=0
for num in arr:
    total=total+num
print("The total sum of the array give is ",total)
```

## Maximunm and minimum element in the array

```python
arr=[1,2,3,40,89,57,5]
max=arr[0]
min=arr[0]
n=len(arr)
for i in range(1,n):
    if arr[i] > max: ## finding the maximum element
        max=arr[i]
    if arr[i] < min: ## finding the minimum elemnet
        min=arr[i]
print("maxixmum value is",max )
print("Minimum value is :",min)
```
