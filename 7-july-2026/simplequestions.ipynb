## Factorial of the Number

example that nu=5 then factorial of the number is 1*2*3*4*5


```python
fact=1
num=int(input("Enter the Number:"))

if num<0:
    print("factorial doesnt exit for the negative number")
elif num==0:
    print("Factorial of 0 is 1")
else:
    for i in range(1,num+1):
        fact=fact * i

    print(f"The factorial of number {num} is {fact}")
```

```python
def factorial(n):
    return 1 if(n==1 or n==0) else n * factorial(n-1)
n=5
print("The factorial of the number",n ,"is ", factorial(n))
```

```python
## Recrsive fuction
def factorial(num):
    if (num==0 or num==1):
        return 1
    else:
        return num * factorial(num-1)
    
num=5
print(factorial(num))
```

## Fibbonic series

A series of the number in which the each number is the sum of the two preceding numbers

```python
n1=0
n2=1
print(n1)
print(n2)

for i in range(2,10):
    sum=n1+n2
    print(sum)
    n1=n2
    n2=sum
```
