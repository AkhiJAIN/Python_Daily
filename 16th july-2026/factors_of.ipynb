## Factores of the given number

##bruteforce solution

```python
num=20
result=[]
for i in range(1,num+1):
    if num % i ==0:
        result.append(i)
print(result)
```

```python
def factore(num):
    result=[]
    for i in range(1,num+1):
        if num % i ==0:
            result.append(i)
    return result
print("The factores of the given number are:",factore(100))
```

```python
num=4
result=[]
for i in range(1,num//2 +1):
    if num % i==0:
        result.append(i)
result.append(num)
print(result)

## Time complexity is o(n/2) ~ o(n)
```
