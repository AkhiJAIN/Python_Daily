## Finding the count of same number how many times it has been repeted

```python
array=[10,20,30,40,50,50,50,60,50,70,60,50,40,50,50,60,51,20,50]
target=50
count=0
for i in array:
    if i==target:
        count=count+1
print(f"{target} has been appeared for the {count} times in the given input")
```

```python
array=[10,20,30,40,50,50,50,60,50,70,60,50,40,50,50,60,51,20,50]
target=20
print(f"{target} has been appeared for the {array.count(target)} times in the given input")
```
