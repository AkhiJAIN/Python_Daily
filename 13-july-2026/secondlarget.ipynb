## Finding the second larget element

```python
def secondlargest(nums):
    lar=second=float('-inf')
    for num in nums:
        if num > lar:
            lar=num
        elif lar > num >second:
            second=num
    return second
nums=[2,6,8,7,2]
print("The second largets elemnt is :",secondlargest(nums))
```
